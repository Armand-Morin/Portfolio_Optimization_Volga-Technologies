```python
from dash import *
import pandas as pd
import yfinance as yf
from jupyter_dash import JupyterDash

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = JupyterDash(__name__,external_stylesheets=external_stylesheets)

app.layout = html.Div([
    dcc.Tabs(id="tabs", value='tab-1', children=[
        dcc.Tab(label='Tab one', value='tab-1'),
        dcc.Tab(label='Tab two', value='tab-2'),]),
    html.Div(id='tabs-content')])


@app.callback(Output('tabs-content', 'children'),
              Input('tabs', 'value'))
def render_content(tab):
    if tab == 'tab-1':
        return html.Div([html.H1(children='Portfolio Management CentraleSupelec'),
                    html.Label('Market to compare performances to'),
                    dcc.Checklist([ '^GSPC', #S&P 500
                         '^DJI', # Dow 30
                         '^IXIC', # Nasdaq
                         '^FCHI', # CAC40
                       ],['^GSPC'],id = 'market_index'
                    ),
                    html.Label('Stocks equity'),
                    dcc.Checklist(['SPY', "QQQ", "VGK", "SCZ", "EWJ", "EEM"],
                                  value = ['SPY', "QQQ", "VGK", "SCZ", "EWJ", "EEM"],
                                  id = 'eq_index'
                    ),
                    html.Label('Stocks finance'),
                    dcc.Checklist(['TIP', "IEF", "TLT", "BWX"],
                                  value = ['TIP', "IEF", "TLT", "BWX"],
                                  id = 'fin_index'
                    ),
                    html.Label('Stocks comodity'),
                    dcc.Checklist(['DBC', "GLD"],
                                  value = ['DBC', "GLD"],
                                  id = 'com_index'
                    ),
                    html.Div(id='my-div_com'),
                    html.Label('Stocks real estate'),
                    dcc.Checklist(['REM', "VNQ", "RWX"],
                                  value = ['REM', "VNQ", "RWX"],
                                  id = 're_index'),], 
                             style={'padding': 20, 'flex': 10, 
                                    'display': 'flex', 'flex-direction': 'column'})
    
    elif tab == 'tab-2':
        return html.Div([
        dcc.Graph(id='stock_price'),
        dcc.Input(id='my-id', value='Prop', type='text'),
        html.Div(id='my-div')
        ],
        style={'display': 'inline-block', 'width': '49%'})

@app.callback(
    Output(component_id='my-div_com', component_property='children'),
    [Input(component_id='com_index', component_property='value')])

def update_output_div(input_value):
    return 'You\'ve entered "{}"'.format(input_value)

if __name__ == '__main__':
    app.run_server(debug = True, mode = 'inline')
```
